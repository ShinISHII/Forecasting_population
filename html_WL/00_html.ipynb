## Compose as pickle

### arrangeTable(df)
縦軸に日付，横軸に時間があり，その交差に水位が表記されたリストを渡した時，

時刻列と水位列のdataframeを返す．

```python
def arrangeTable(df):
    df=df.drop(df.index[[0,1]])     #最初2行を削除
    df=df.drop(df.index[len(df)-1]) #最後1行を削除
    df=df.set_index(df.columns[0])  #indexをstr(datetime)に変更

    datetimeList=[] #pd.datetime型の日時リスト
    valueList=[] #水位リスト

    #各月毎にループする．
    for index,row in df.iterrows(): 
        for col in df.columns:
            # print(f"Index: {index}, Column: {col}, Value: {value}")

            # 水位リスト作成
            value=row[col]
            valueList.append(value)

            #日時リスト作成
            datetime_Ymd = pd.to_datetime(index, format='%Y/%m/%d', errors='coerce')
            if col==24:
                col=0
                datetime_Ymd = datetime_Ymd + pd.Timedelta(days=1)
            result = datetime_Ymd + pd.to_timedelta(col,unit="h")
            datetimeList.append(result)


    # print(datetimeList)
    # print(valueList)
            
    df_new=pd.DataFrame({
        "datetime":datetimeList,
        "waterLevel":valueList
    })
    df_new = df_new.set_index('datetime')

    return df_new
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from io import StringIO 

# データ格納用リスト
df_list = []

# 開始日と終了日を設定
start_date = pd.to_datetime('2016-01-01')
end_date = pd.to_datetime('2024-12-01')

# 現在の日付を開始日で設定
current_date = start_date

while current_date <= end_date:
    # URLの構築
    url = f'http://www1.river.go.jp/cgi-bin/DspWaterData.exe?KIND=2&ID=307111287708010&BGNDATE={current_date.strftime("%Y%m%d")}&ENDDATE=20241231&KAWABOU=NO'

    # HTMLコンテンツの取得
    response = requests.get(url)

    # BeautifulSoupでHTMLを解析
    soup = BeautifulSoup(response.content, 'html.parser')

    # pandasでHTMLテーブルを抽出
    tables = pd.read_html(StringIO(str(soup)))

    # 最初のテーブルを取得（インデックスを1に指定）
    df = tables[1]

    # テーブル整理
    df = arrangeTable(df)

    # リストに追加
    df_list.append(df)

    # 次の月に進む
    current_date += pd.DateOffset(months=1)

# 最後に、データフレームリストを結合
df_wl = pd.concat(df_list, ignore_index=False)
df_wl.to_pickle('./kurashiki/waterLevel.pickle')
# 結果の表示（任意）
# print(df_wl)
```

## Open Pickle

```python
import pandas as pd

# pickleファイルからDataFrameを読み込む
df_wl = pd.read_pickle('./kurashiki/waterLevel.pickle')

# 読み込んだDataFrameを表示（任意）
df_wl
## 日本語（'閉局'）やNaNが入っている
```

```python
# pd_wlのデータを線形補完する．

# Replace Japanese characters with NaN
df_wl['waterLevel'] = pd.to_numeric(df_wl['waterLevel'], errors='coerce')

# Use liner interpolation to fill NaN values.
# df_wl['waterLevel'] = df_wl['waterLevel'].interpolate(method='ffill')
df_wl['waterLevel'] = df_wl['waterLevel'].interpolate(method='linear')
print(df_wl)
```
