## Save HTML as csv

```python
import pandas as pd
import os

# データを保存するフォルダ
save_folder = '../html/kurashiki/raw'
os.makedirs(save_folder, exist_ok=True)

# ページ数に基づきfor文を使ってループ
for page in range(1, 32):  # ページ数1〜47
    url = f'http://agora.ex.nii.ac.jp/cgi-bin/cps/warning_list.pl?acode=3320200&page={page}'
    
    try:
        df = pd.read_html(url)[0]  # HTMLからデータを取得
        #ページ名を2桁表示
        page_str = '{:02}'.format(page)
        file_name = f'warning_kurashiki_{page_str}.csv'  # 各ページごとのファイル名
        file_path = os.path.join(save_folder, file_name)
        
        df.to_csv(file_path, index=False)  # CSVファイルとして保存
        print(f"Saved data from page {page_str} to {file_name}")
    
    except Exception as e:
        print(f"Failed to retrieve data from page {page_str}: {e}")

# 保存したCSVファイルを後で読み込むとき
# all_data = pd.concat([pd.read_csv(os.path.join(save_folder, f)) for f in os.listdir(save_folder)])
```
